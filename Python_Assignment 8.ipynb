```python
from sklearn.datasets import load_iris
import pandas as pd
import matplotlib.pyplot as plt        
from mpl_toolkits.mplot3d import Axes3D
```

```python
iris = load_iris()
x = iris.data
y = iris.target
columns = iris.feature_names
trg = iris.target_names
```

```python
df = pd.DataFrame(data = x ,columns=columns)
df['Target'] = y
```

```python
def plot_iris(i,j,k):
    x = df[df['Target'] == 2].iloc[:,i]
    y = df[df['Target'] == 2].iloc[:,j]
    z = df[df['Target'] == 2].iloc[:,k]

    x2 = df[df['Target'] == 1].iloc[:,i]
    y2 = df[df['Target'] == 1].iloc[:,j]
    z2 = df[df['Target'] == 1].iloc[:,k]

    x3 = df[df['Target'] == 0].iloc[:,i]
    y3 = df[df['Target'] == 0].iloc[:,j]
    z3 = df[df['Target'] == 0].iloc[:,k]
    
    ax = plt.figure(figsize = (12,12)).gca(projection='3d')
    ax.scatter(x,y,z,color = 'b' ,alpha =0.9,s=40,label = trg[2])
    ax.scatter(x2,y2,z2,color = 'g',alpha = 0.9,s=40,label = trg[1])
    ax.scatter(x3,y3,z3,color = 'r',alpha = 0.9,s=40,label = trg[0])
    ax.set_xlabel(df.columns[i])
    ax.set_ylabel(df.columns[j])
    ax.set_zlabel(df.columns[k])
    plt.show()
```

```python
#Taking column 0 , 1, 2
plot_iris(0,1,2)
```

```python
#Taking column 0 , 1, 3
plot_iris(0,1,3)
```

```python
#Taking column 0 , 2, 3
plot_iris(0,2,3)
```

```python
#Taking column  1, 2,3
plot_iris(1,2,3)
```
